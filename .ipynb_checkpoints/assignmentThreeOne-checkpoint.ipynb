```python
import numpy as np
from qiskit import QuantumCircuit

#1.2 QFT acting on the first n qubits
def apply_qft(n, circuit):
    """
    Applies QFT to the first n qubits of the circuit.
    """
    # 1. Apply Hadamard and Controlled Phase gates
    # Iterate j from n-1 down to 0
    for j in range(n - 1, -1, -1):
        circuit.h(j)
        # Iterate k from j-1 down to 0
        for k in range(j - 1, -1, -1):
            # Phase shift theta = pi / 2^(j-k)
            theta = np.pi / (2 ** (j - k))
            circuit.cp(theta, k, j) # Control: k, Target: j
        circuit.barrier()

    # 2. Swap qubits to reverse the order
    for i in range(n // 2):
        circuit.swap(i, n - i - 1)
    
    circuit.barrier()

#1.3 Inverse QFT
def apply_iqft(n, circuit):
    """
    Applies Inverse QFT to the first n qubits.
    Reverses the order of gates and uses negative phases.
    """
    # 1. Undo Swaps (Inverse of Swap is Swap)
    for i in range(n // 2):
        circuit.swap(i, n - i - 1)
        
    circuit.barrier()
        
    # 2. Undo Gates (Reverse order: j from 0 to n-1)
    for j in range(n):
        # Undo Controlled Phase gates (Reverse k order too: 0 to j-1)
        # Inverse of CP(theta) is CP(-theta)
        for k in range(j):
            theta = -np.pi / (2 ** (j - k))
            circuit.cp(theta, k, j)
        circuit.h(j)
        circuit.barrier()

#1.4 Generalizing the functions 
def apply_qft_gen(qubits, circuit):
    """
    Applies QFT to the specific qubits listed in `qubits`.
    qubits[0] corresponds to index 0, etc.
    """
    n = len(qubits)
    
    # 1. Gates
    for j in range(n - 1, -1, -1):
        target_qubit = qubits[j]
        circuit.h(target_qubit)
        
        for k in range(j - 1, -1, -1):
            control_qubit = qubits[k]
            theta = np.pi / (2 ** (j - k))
            circuit.cp(theta, control_qubit, target_qubit)
            
    # 2. Swaps
    for i in range(n // 2):
        circuit.swap(qubits[i], qubits[n - i - 1])

def apply_iqft_gen(qubits, circuit):
    """
    Applies Inverse QFT to the specific qubits listed in `qubits`.
    """
    n = len(qubits)
    
    # 1. Swaps
    for i in range(n // 2):
        circuit.swap(qubits[i], qubits[n - i - 1])
        
    # 2. Gates
    for j in range(n):
        target_qubit = qubits[j]
        for k in range(j):
            control_qubit = qubits[k]
            theta = -np.pi / (2 ** (j - k))
            circuit.cp(theta, control_qubit, target_qubit)
        circuit.h(target_qubit)

# Print of QFT acting on the first n qbits
print("--- QFT n=3 ---")
n = 3
qc3 = QuantumCircuit(n)
apply_qft(n, qc3)
print(qc3)

# Print of Inverse QFT
print("--- Result for Task 1.2 (n=4) ---")
n = 4
qc4 = QuantumCircuit(n, 0)
apply_qft(n, qc4)
qc4.draw('mpl')
```
